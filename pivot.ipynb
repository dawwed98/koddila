```python
import pandas as pd
df = pd.read_excel('Pivot.xlsx')
df
```

```python
import numpy as np
df.pivot_table(values='Sprzedaż', index='Przedstawiciel', columns='Region', aggfunc=np.sum)
```

```python
df.pivot_table(values='Sprzedaż', index=['Region', 'Przedstawiciel'], aggfunc=np.sum).round(2)
```

```python
df.pivot_table(values='Sprzedaż', index='Region', aggfunc=[len, np.max, np.min]).round(0)
```

```python
def commission_fee(x):
    if x <= 300:
        return 0
    elif x <= 900:
        return x * 0.03
    else:
        return x * 0.06
df['Prowizja'] = df['Sprzedaż'].apply(lambda x: commission_fee(x)).round(2)
df
```
