```python
import numpay as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import sys
```

```python
# Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami i wynik swojej pracy wyślij
import numpy as np

średnia = 3
odchylenie_standardowe = 1
n = 100
n2 = 10000
# Generowanie danych
dane1 = np.random.normal(średnia, odchylenie_standardowe, n)
dane2 = np.random.normal(średnia, odchylenie_standardowe, n2)

# Obliczanie statystyk opisowych
średnia1 = np.mean(dane1)
średnia2 = np.mean(dane2)
mediana1 = np.median(dane1)
mediana2 = np.median(dane2)
wariancja1 = np.var(dane1, ddof=1)
wariancja2 = np.var(dane2, ddof=1)
odchylenie_standardowe1 = np.std(dane1, ddof=1)
odchylenie_standardowe2 = np.std(dane2, ddof=1)
skew1 = pd.Series(dane1).skew()
skew2 = pd.Series(dane2).skew()
kurtoza1 = pd.Series(dane1).kurtosis()
kurtoza2 = pd.Series(dane2).kurtosis()

# Tworzenie DataFrame do przechowywania wyników
statystyki = pd.DataFrame({
    'Statystyka': ['Średnia', 'Mediana', 'Wariancja', 'Odchylenie standardowe', 'Skewness', 'Kurtoza'],
    'Zbiór 1 (n=100)': [średnia1, mediana1, wariancja1, odchylenie_standardowe1, skew1, kurtoza1],
    'Zbiór 2 (n=10000)': [średnia2, mediana2, wariancja2, odchylenie_standardowe2, skew2, kurtoza2]
})

# Wyświetlanie wyników
print(statystyki)
# Tworzenie wykresów
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(dane1, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram dla zbioru 1 (n=100)')
plt.xlabel('Wartości')
plt.ylabel('Częstość')
plt.subplot(1, 2, 2)
plt.hist(dane2, bins=30, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram dla zbioru 2 (n=10000)')
plt.xlabel('Wartości')
```
