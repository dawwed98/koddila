```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('HRDataset.csv')
df
```

```python
# Rozkad wynikw wydajnoci w zalenoci od menedera
sns.displot(df, x='PerformanceScore', y='ManagerName', hue='PerformanceScore', palette='Set2')
plt.title('Rozkad wynikw wydajnoci w zalenoci od menedera')
plt.xlabel('Wynik wydajnoci')
plt.ylabel('Meneder')
plt.show()

```

```python
# Jakie rda pozyskania pracownika (Recruitment Source) s najlepsze, jeli zaley nam na jak najduszym stau pracownikw?

# Wykres pudekowy dla dugoci stau w zalenoci od rda pozyskania
# Obliczenie dugoci stau w latach
df['LengthofService'] = pd.to_datetime('today').year - pd.to_datetime(df['DateofHire']).dt.year

# Wywietlenie wykresu pudekowego dla dugoci stau w zalenoci od rda pozyskania
sns.boxplot(x='RecruitmentSource', y='LengthofService', data=df, palette='Set2')
plt.xticks(rotation=45, ha='right') # Rotacja etykiet na osi x
plt.tight_layout() # Dostosowanie ukadu wykresu
plt.grid(axis='y') # Dodanie siatki tylko na osi y
plt.title('Dugo stau pracownikw w zalenoci od rda pozyskania') 
plt.xlabel('rdo pozyskania')
plt.ylabel('Dugo stau (lata)') 
plt.show()
```

```python
# Czy stan cywilny (MartialDesc) pracownika koreluje w jakikolwiek sposb z zadowoleniem z pracy (EmpSatisfaction)?

# Wykres pudekowy dla zadowolenia z pracy w zalenoci od stanu cywilnego
sns.boxplot(x='MaritalDesc', y='EmpSatisfaction', data=df, palette='Set1')
plt.xticks(rotation=45, ha='right') # Rotacja etykiet na osi x
plt.tight_layout() # Dostosowanie ukadu wykresu
```

```python
# Jak wyglda struktura wieku aktualnie zatrudnionych pracownikw?

# Obliczenie wieku pracownikw
df['DOB'] = pd.to_datetime(df['DOB'], format='%d/%m/%y', errors='coerce', dayfirst=True)
df = df.dropna(subset=['DOB'])  # Usunicie wierszy z nieprawidowymi datami urodzenia
df['Age'] = pd.to_datetime('today').year - df['DOB'].dt.year

# print(df['DOB'].unique()) # Wywietlenie unikalnych wartoci daty urodzenia

# Wywietlenie histogramu wieku pracownikw
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Rozkad wieku pracownikw')
plt.xlabel('Wiek')
plt.xlim(0, 60) # Ustawienie limitu osi x
plt.ylabel('Liczba pracownikw')
plt.grid(axis='x') # Dodanie siatki tylko na osi x
plt.show()
```

```python
#Czy starsi pracownicy pracuj nad wiksz liczb specjalnych projektw ni modsi pracownicy?

# Wykres pudekowy dla liczby projektw w zalenoci od wieku
sns.boxplot(x='Age', y='SpecialProjectsCount', data=df, palette='Set2')
plt.xticks(rotation=60, ha='right') # Rotacja etykiet na osi x
plt.grid(axis='y') # Dodanie siatki tylko na osi y
plt.title('Liczba projektw w zalenoci od wieku pracownikw')
plt.xlabel('Wiek')
plt.ylabel('Liczba projektw')
plt.tight_layout() # Dostosowanie ukadu wykresu
plt.show()
```
