```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler

dataset = pd.read_csv('titanic.csv')
dataset
```

```python
fare_values = dataset.loc[dataset['Fare'].notnull(),'Fare'].values

```

```python
# obliczenie średniej, kwartyli

mean = np.mean(fare_values).round(2)
median = np.median(fare_values)

q1 = np.percentile(fare_values, 25)
q2 = np.percentile(fare_values, 50)
q3 = np.percentile(fare_values, 75)
q4 = np.percentile(fare_values, 100)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Q1: {q1}")
print(f"Q2: {q2}")
print(f"Q3: {q3}")
print(f"Q4: {q4}")

plt.boxplot(fare_values, vert=False)
plt.title('Wykres pudełkowy wartości Fary')
plt.xlabel('Fare')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(median, color='blue', linestyle='--', label='Median')
plt.axvline(q1, color='green', linestyle='--', label='Q1')
plt.axvline(q4, color='orange', linestyle='--', label='Q4')
plt.legend(loc='upper right')
plt.show()
```
