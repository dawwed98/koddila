```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('HRDataset.csv')
df
```

```python
# Rozkład wyników wydajności w zależności od menedżera
sns.displot(df, x='PerformanceScore', y='ManagerName', hue='PerformanceScore', palette='Set2')
plt.title('Rozkład wyników wydajności w zależności od menedżera')
plt.xlabel('Wynik wydajności')
plt.ylabel('Menedżer')
plt.show()

```

```python
# Jakie źródła pozyskania pracownika (Recruitment Source) są najlepsze, jeśli zależy nam na jak najdłuższym stażu pracowników?

# Wykres pudełkowy dla długości stażu w zależności od źródła pozyskania
# Obliczenie długości stażu w latach
df['LengthofService'] = pd.to_datetime('today').year - pd.to_datetime(df['DateofHire']).dt.year

# Wyświetlenie wykresu pudełkowego dla długości stażu w zależności od źródła pozyskania
sns.boxplot(x='RecruitmentSource', y='LengthofService', data=df, palette='Set2')
plt.xticks(rotation=45, ha='right') # Rotacja etykiet na osi x
plt.tight_layout() # Dostosowanie układu wykresu
plt.grid(axis='y') # Dodanie siatki tylko na osi y
plt.title('Długość stażu pracowników w zależności od źródła pozyskania') 
plt.xlabel('Źródło pozyskania')
plt.ylabel('Długość stażu (lata)') 
plt.show()
```

```python
# Czy stan cywilny (MartialDesc) pracownika koreluje w jakikolwiek sposób z zadowoleniem z pracy (EmpSatisfaction)?

# Wykres pudełkowy dla zadowolenia z pracy w zależności od stanu cywilnego
sns.boxplot(x='MaritalDesc', y='EmpSatisfaction', data=df, palette='Set1')
plt.xticks(rotation=45, ha='right') # Rotacja etykiet na osi x
plt.tight_layout() # Dostosowanie układu wykresu
```

```python
# Jak wygląda struktura wieku aktualnie zatrudnionych pracowników?

# Obliczenie wieku pracowników
df['DOB'] = pd.to_datetime(df['DOB'], format='%d/%m/%y', errors='coerce', dayfirst=True)
df = df.dropna(subset=['DOB'])  # Usunięcie wierszy z nieprawidłowymi datami urodzenia
df['Age'] = pd.to_datetime('today').year - df['DOB'].dt.year

# print(df['DOB'].unique()) # Wyświetlenie unikalnych wartości daty urodzenia

# Wyświetlenie histogramu wieku pracowników
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Rozkład wieku pracowników')
plt.xlabel('Wiek')
plt.xlim(0, 60) # Ustawienie limitu osi x
plt.ylabel('Liczba pracowników')
plt.grid(axis='x') # Dodanie siatki tylko na osi x
plt.show()
```

```python
#Czy starsi pracownicy pracują nad większą liczbą specjalnych projektów niż młodsi pracownicy?

# Wykres pudełkowy dla liczby projektów w zależności od wieku
sns.boxplot(x='Age', y='SpecialProjectsCount', data=df, palette='Set2')
plt.xticks(rotation=60, ha='right') # Rotacja etykiet na osi x
plt.grid(axis='y') # Dodanie siatki tylko na osi y
plt.title('Liczba projektów w zależności od wieku pracowników')
plt.xlabel('Wiek')
plt.ylabel('Liczba projektów')
plt.tight_layout() # Dostosowanie układu wykresu
plt.show()
```
