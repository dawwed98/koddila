```python
# Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkam
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
# Tworzenie danych z rozkładu normalnego
data_100 = np.random.normal(3, 1, 100)
data_10000 = np.random.normal(3, 1, 10000)
# Tworzenie DataFrame
df_100 = pd.DataFrame(data_100, columns=['data'])
df_10000 = pd.DataFrame(data_10000, columns=['data'])
# Obliczanie statystyk opisowych
stats_100 = df_100.describe()
stats_10000 = df_10000.describe()
# Wyświetlanie statystyk
print("Statystyki dla 100 próbek:")
print(stats_100)
print("\nStatystyki dla 10 000 próbek:")
print(stats_10000)
# Wizualizacja danych
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data_100, bins=20, color='blue', alpha=0.7, label='100 próbek')
plt.title('Histogram dla 100 próbek')
plt.xlabel('Wartości')
plt.ylabel('Częstość')
plt.legend()
plt.subplot(1, 2, 2)
plt.hist(data_10000, bins=20, color='green', alpha=0.7, label='10 000 próbek')
plt.title('Histogram dla 10 000 próbek')
plt.xlabel('Wartości')
plt.ylabel('Częstość')
plt.legend()
plt.tight_layout()
plt.show()
# Opis różnic pomiędzy statystykami
print("\nRóżnice pomiędzy statystykami:")
print("W przypadku 100 próbek średnia może być bardziej podatna na wpływ wartości odstających, podczas gdy w przypadku 10 000 próbek średnia będzie bardziej stabilna.")
print("2.Przy mniejszej liczbie próbek odchylenie standardowe może być większe, co oznacza większą zmienność danych. W przypadku większej liczby próbek odchylenie standardowe będzie bardziej zbliżone do rzeczywistego odchylenia standardowego rozkładu.")



```
